## Visualisation

### Top 10 Product ids

### Checking for the rating population in our dataset

```python
pie = df.rating.value_counts()                # For the counts of different categegories in overall
print(pie)
plt.figure(figsize=(14,6))                     # Set the figure size
plt.title('Ratings')                   # Title of plot
plt.pie(pie,labels=df['rating'].unique(),autopct = '%0.1f%%',shadow=True)     # auto pct is giving percentage of total
plt.plot();
```

#### Observation: From the above pie chart we can see that our dataset have 66.4% of (2 rating), 5.2% of (3 rating), 7.3% of (1 rating), 8.9% of (4 rating), 12.3% of (5 rating).

### Price Distribution

```python
sns.boxplot(df['price'],showmeans=True,color='yellow');
plt.title('Distribution of Price');
```

#### Observation: Here we can see that 50% of our data's price range lies between 20 to 40.As we can see that the mean is greater than the median so it's right skewed

### Top 10 Titles as per the ratings

```python
x=df.groupby('title')['rating'].mean()
x=x.sort_values(ascending=False)
x
```

```python
a=pd.DataFrame({'TOP_10_titles':(x[:10])})
a.plot.barh(color='red')
plt.xlabel('rating')
plt.plot();
```

#### Observation: From the above graph we can see the top10 titles as per the ratings to them

### Least 10 Titles as per the ratings

```python
y=df.groupby('title')['rating'].mean()
y=y.sort_values(ascending=True)
y
```

```python
a=pd.DataFrame({'Least_10_titles':(y[:10])})
a.plot.barh(color='red')
plt.xlabel('rating')
plt.plot();
```

#### Observation: From the above graph we can see the Bottom10 titles as per the ratings to them

### Least 10 titles as per rank

```python
y=df.groupby('title')['rank_no'].mean()
y=y.sort_values(ascending=True)
y
```

```python
a=pd.DataFrame({'Least_10_titles_as_per_rank':(y[:10])})
a.plot.barh(color='red')
plt.xlabel('rank')
plt.plot();
```

#### Observation: From the above graph we can see that 'Olaplex Hair perfector' is the title with lowest rank

### Top 10 title as per the rank no

```python
x=df.groupby('title')['rank_no'].mean()
x=x.sort_values(ascending=False)
x
```

```python
a=pd.DataFrame({'Top_10_titles_as_per_rank':(x[:10])})
a.sort_values('Top_10_titles_as_per_rank').plot.barh(color='red')
plt.xlabel('rank')
plt.plot();
```

#### Observation: From the above graph we can see that 'AHAVA 3 X The Charm' is the title with highest rank

### Price as per years

```python
df1 = df.set_index('date')      
df2=df1.resample('y').mean()
```

```python
plt.figure(figsize=(20,7))
df2['price'].plot()
plt.title('Prices variation over the years')
plt.plot()
```

#### Observation: Here we can see that the price is decreasing over the years
